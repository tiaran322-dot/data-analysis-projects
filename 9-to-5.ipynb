## Aggregate Your Friends and Code some SQL

Put together the following:

1. How many employees were reported in 2016 in all industries? Round to the nearest whole number.

```sql
-- There were 2340612 total employees in all industries in 2016.
SELECT Round(SUM(a.value),0) as total_employees, s.series_title
FROM LaborStatisticsDB.dbo.series as s
INNER JOIN LaborStatisticsDB.dbo.annual_2016 as a ON s.series_id = a.series_id
WHERE data_type_code = 1
GROUP BY s.series_title
```

2. How many women employees were reported in 2016 in all industries? Round to the nearest whole number. 

```sql
--There are 1125490 total women in all industries in 2016
SELECT Round(SUM(a.value),0) as total_employees, s.series_title
FROM LaborStatisticsDB.dbo.series as s
INNER JOIN LaborStatisticsDB.dbo.annual_2016 as a ON s.series_id = a.series_id
WHERE data_type_code = 10
GROUP BY s.series_title
```

3. How many production/nonsupervisory employees were reported in 2016? Round to the nearest whole number. 

```sql
--There are 1263650 total production/nonsupervisory employees in 2016.
SELECT Round(SUM(a.value),0) as total_employees, s.series_title
FROM LaborStatisticsDB.dbo.series as s
INNER JOIN LaborStatisticsDB.dbo.annual_2016 as a ON s.series_id = a.series_id
WHERE data_type_code = 6
GROUP BY s.series_title
```

4. In January 2017, what is the average weekly hours worked by production and nonsupervisory employees across all industries?

```sql
-- The average weekly hours worked by production and nonsupervisory is 36.06.
SELECT Round(AVG(j.value), 2) as avg_weekly_hours, s.series_title
FROM LaborStatisticsDB.dbo.series as s
INNER JOIN LaborStatisticsDB.dbo.january_2017 as j ON s.series_id = j.series_id
WHERE data_type_code = 7
GROUP BY s.series_title
```

5. What is the total weekly payroll for production and nonsupervisory employees across all industries in January 2017? Round to the nearest penny.

```sql
-- The total weekly payroll is $1838753220 for production/nonsup employees across all industries.
SELECT Round(SUM(j.value), 2) as total_weekly_payroll, s.series_title
FROM LaborStatisticsDB.dbo.series as s
INNER JOIN LaborStatisticsDB.dbo.january_2017 as j ON s.series_id = j.series_id
WHERE data_type_code = 82
GROUP BY s.series_title
```

6. In January 2017, for which industry was the average weekly hours worked by production and nonsupervisory employees the highest? Which industry was the lowest?

```sql
-- The Motor vehicle power train components industry has the highest average hours worked.
SELECT Round(AVG(j.value), 2) as avg_weekly_hours, i.industry_name, s.series_title
FROM LaborStatisticsDB.dbo.series as s
INNER JOIN LaborStatisticsDB.dbo.january_2017 as j ON s.series_id = j.series_id
INNER JOIN LaborStatisticsDB.dbo.industry as i on s.industry_code = i.industry_code
WHERE data_type_code = 7
GROUP BY s.series_title, j.period, i.industry_name
ORDER BY avg_weekly_hours DESC
```

7. In January 2021, for which industry was the total weekly payroll for production and nonsupervisory employees the highest? Which industry was the lowest?

```sql
--The private sector is the highest and the Coin-operated laundries and drycleaners is the lowest.
SELECT Round(SUM(j.value), 2) as total_weekly_payroll, i.industry_name, s.series_title
FROM LaborStatisticsDB.dbo.series as s
INNER JOIN LaborStatisticsDB.dbo.january_2017 as j ON s.series_id = j.series_id
INNER JOIN LaborStatisticsDB.dbo.industry as i on s.industry_code = i.industry_code
WHERE data_type_code = 82
GROUP BY s.series_title, j.period, i.industry_name
ORDER BY total_weekly_payroll DESC
```

## Subqueries, Unions, Derived Tables, Oh My!

1. Write a query that returns the `series_id`, `industry_code`, `industry_name`, and `value` from the `january_2017` table but only if that value is greater than the average value for `annual_2016` of `data_type_code` 82.

```sql
SELECT j.series_id, i.industry_code, i.industry_name, j.value
FROM LaborStatisticsDB.dbo.january_2017 as j
INNER JOIN LaborStatisticsDB.dbo.series as s ON s.series_id = j.series_id
INNER JOIN LaborStatisticsDB.dbo.industry as i ON j.id = i.id
WHERE j.value > (SELECT AVG(value) FROM LaborStatisticsDB.dbo.annual_2016) AND s.data_type_code = 82
```

```sql
SELECT a.year, a.period, Round(AVG(a.value),2) as avg_weekly_earnings 
FROM LaborStatisticsDB.dbo.annual_2016 as a
INNER JOIN LaborStatisticsDB.dbo.series as s ON a.series_id = s.series_id
WHERE s.data_type_code = 30
GROUP BY a.year, a.period

UNION
    
SELECT j.year, j.period, Round(AVG(j.value),2) as avg_weekly_earnings
FROM LaborStatisticsDB.dbo.january_2017 as j
INNER JOIN LaborStatisticsDB.dbo.series as s ON j.series_id = s.series_id
WHERE s.data_type_code = 30
GROUP BY j.year, j.period
```
